```python
import re

reviews=[

         "Amazing Phone!!!",
         "BAD battery life ",
         "adlivery was late .."

]

cleaned_reviews=[]


for review_text in reviews:
    # Convert to lowercase and remove non-alphanumeric characters
    cleaned_text = re.sub("[^a-zA-Z0-9]"," ", review_text.lower())
    # Remove leading/trailing whitespace
    cleaned_text = cleaned_text.strip()
    cleaned_reviews.append(cleaned_text)

print(cleaned_reviews)
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd

data={
      "hours_studies":[1,2,3,4,5,6,7,8],
      "pass":[0,0,0,1,1,1,1,1]
}

df = pd.DataFrame(data)

x=df[["hours_studies"]]
y=df["pass"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(
    x,y,test_size=0.25,
    random_state=42)

print("Training Data ", x_train)

print("Testing Data ", x_test)
```

```python
print("Testing Data ", y_test)
```

### Create and Save a new CSV Dataset

```python
import pandas as pd
import numpy as np

# Generate dummy data for 10 rows
hours_stud = np.random.randint(1, 10, size=10) # Random hours between 1 and 9
test_grade = np.random.randint(50, 100, size=10) # Random grades between 50 and 99

# Create a dictionary to hold the data
data = {
    'hours_stud': hours_stud,
    'test_grade': test_grade
}

# Create a DataFrame
df_new = pd.DataFrame(data)

# Save the DataFrame to a CSV file
csv_file_path = 'student_data.csv'
df_new.to_csv(csv_file_path, index=False)

print(f"CSV file '{csv_file_path}' created successfully.")

# Display the first 5 rows of the new DataFrame
display(df_new.head())
```
