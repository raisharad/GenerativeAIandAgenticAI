```python
from google.colab import ai
response = ai.generate_text("What is the capital of France?")

import numpy as np
import time

# Define Matrix A
A = np.array([
    [1, 2, 4],
    [3, 4, 6],
     [4, 1, 9]
])

# Define Matrix B
B = np.array([
    [5, 6,9],
    [7, 8,1],
     [5, 1,7]

])


start = time.time()
# Multiply matrices
C = np.matmul(A, B)

# Print results
print("Matrix A:")
print(A)

print("\nMatrix B:")
print(B)

print("\nMatrix Multiplication Result (A x B):")
print(C)

end = time.time()
print("cpu time" , end-start)




```

```python
import torch
import time

# Use GPU
device = "cuda"

start = time.time()

# Create random matrices
A = torch.rand(2000, 2000).to(device)
B = torch.rand(2000, 2000).to(device)

# Matrix multiplication
C = torch.matmul(A, B)

# Wait for GPU operations to finish
#torch.cuda.synchronize()

end = time.time()

print("GPU Time:", end - start)
```

```python
from transformers import pipeline
import torch

# Load GPT-2 text generation pipeline
generator = pipeline("text-generation", model="gpt2")

# Prompt
prompt = "Artificial intelligence will transform healthcare by"

# Generate text
result = generator(
    prompt,
    max_length=50,
    num_return_sequences=1
)

# Print output
print(result[0]["generated_text"])
```
