```python
import pandas as pd

data={

      "customer_id" :[101,102,103],
      "name" : ["alice",  "rahul", "arjun"],
      'balance':[5000,70000,23000] ,
      'loan_status': ["approved", "rejected", "approved"]
}

df = pd.DataFrame(data)
print(df)
approved=df[df["loan_status"] == "approved"]
print(approved)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

documents={

          "AI is poerful",
          "AI is transforming healthcare"
}

vertorizer=CountVectorizer()

x=vertorizer.fit_transform(documents)
print(vertorizer.get_feature_names_out())
print(x.toarray())
```
