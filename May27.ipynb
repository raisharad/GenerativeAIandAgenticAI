```python
from sklearn.feature_extraction.text import CountVectorizer

documents={

          "AI is poerful",
          "AI is transforming healthcare"
}

vertorizer=CountVectorizer()

x=vertorizer.fit_transform(documents)
print(vertorizer.get_feature_names_out())
print(x.toarray())
```
