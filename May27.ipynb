```python
import pandas as pd

# Create a dictionary with the specified data
data_new = {
    'name': ['rahul', 'priya', 'rahul', 'meera'],
    'age': [25, 100, 25, pd.NA],
    'city': ['bangunalore', 'mumbai', 'bangunalore', 'chennai']
}

# Create the DataFrame
df_new = pd.DataFrame(data_new)

# Display the new DataFrame
print(df_new)
```

```python
df_new= df_new.drop_duplicates()
print(df_new)
```

```python
df_new['age']=df_new['age'].fillna(df_new['age'].mean())
print(df_new)
```

```python
df_new['city']=df_new['city'].str.upper()
print(df_new)
```

```python
df_new['age'] = df_new['age'].where(df_new['age'] < 100, pd.NA)
print(df_new)
```
